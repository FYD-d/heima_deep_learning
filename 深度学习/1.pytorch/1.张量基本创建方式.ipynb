```python
"""
案例：
    演示张量的基本创建方式

张量:
    PyTorch框架属于最常用的深度学习，无论是后需要学习的ANN(人工神经网络)，还是CNN(卷积神经网络)，RNN(循环神经网络)
    底层在处理数据时，都是使用 张量 来处理的

    张量  ——>  存储同一类型元素的容器，且元素值必须是 数值才可以
    张量的基本创建方式:
    torch.tensor 根据指定数据创建张量
    torch.Tensor 根据形状创建张量, 其也可用来创建指定数据的张量
    torch.IntTensor、torch.FloatTensor、torch.DoubleTensor 创建指定类型的张量

细节:
    Tensor方式较之于 tensor方式, 可以基于 形状来直接创建张量.

需要你掌握的方式:
    tensor(值, 类型), 例如:
        data = np.random.randint(0, 10, size=(2, 3))
        t3 = torch.tensor(data, dtype=torch.float)
"""
```

```python
# 导包
import torch
import numpy as np

# 1.定义函数，演示：torch.tensor 根据指定数据创建张量
def dm01():
    # 场景1：标量  张量
    t1=torch.tensor(10)
    print(f't1: {t1} \ntype: {type(t1)}')
    print('-'*30)

    # 场景2：二维列表 ——> 张量
    data=[[1,2,3],[4,5,6]]
    t2=torch.tensor(data, dtype=torch.float)
    print(f't2: {t2} \ntype: {type(t2)}')
    print('-'*30)

    # 场景3：numpy nd数组 ——> 张量
    data=np.random.randint(0,10,size=(2,3))
    t3=torch.tensor(data)
    print(f't3: {t3} \ntype: {type(t3)}')
    print('-'*30)
    
    # 场景4：尝试直接创建 指定维度(例如：2行3列的)张量
    # 报错
    # t4=torch.tensor(2,3)
    # print(f't4: {t4} \ntype: {type(t4)}')
dm01()
```

```python
# 2.定义函数，演示：torch.Tensor 根据形状创建张量, 其也可用来创建指定数据的张量
def dm02():
    # 场景1：标量  张量
    t1=torch.Tensor(10)
    print(f't1: {t1} \ntype: {type(t1)}')
    print('-'*30)

    # 场景2：二维列表 ——> 张量
    data=[[1,2,3],[4,5,6]]
    t2=torch.Tensor(data)
    print(f't2: {t2} \ntype: {type(t2)}')
    print('-'*30)

    # 场景3：numpy nd数组 ——> 张量
    data=np.random.randint(0,10,size=(2,3))
    t3=torch.Tensor(data)
    print(f't3: {t3} \ntype: {type(t3)}')
    print('-'*30)
    
    # 场景4：尝试直接创建 指定维度(例如：2行3列的)张量
    t4=torch.Tensor(2,3)
    print(f't4: {t4} \ntype: {type(t4)}')
dm02()
```

```python
# 3.定义函数，演示：torch.IntTensor、torch.FloatTensor、torch.DoubleTensor 创建指定类型的张量
def dm03():
    # 场景1：标量  张量
    t1=torch.IntTensor(10)
    print(f't1: {t1} \ntype: {type(t1)}')
    print('-'*30)

    # 场景2：二维列表 ——> 张量
    data=[[1,2,3],[4,5,6]]
    t2=torch.FloatTensor(data)
    print(f't2: {t2} \ntype: {type(t2)}')
    print('-'*30)

    # 场景3：numpy nd数组 ——> 张量
    data=np.random.randint(0,10,size=(2,3))
    t3=torch.IntTensor(data)
    print(f't3: {t3} \ntype: {type(t3)}')
    print('-'*30)
    
    # 场景4：如果类型不匹配，会自动尝试转换类型
    data=np.random.randint(0,10,size=(2,3))
    t4=torch.FloatTensor(data)    # 默认：float32
    print(f't4: {t4} \ntype: {type(t4)}')
    print('-'*30)
    
dm03()
```
