```python
"""
案例:
    演示张量和numpy之间如何相互转换，以及 如何从标量张量中提取其内容

涉及到的API：
    场景1：张量 ——> numpy nd数组对象
            张量对象.numpy()           共享内存
            张量对象.numpy().copy()    不共享内存
    场景2：numpy nd数组 ——> 张量
            from_numpy()            共享内存
            torch.tensor(nd数组)     不共享内存
    场景3：从标量中提取其内容
            标量张量.item()

掌握：
    张量 ——> numpy： 张量对象.numpy()
    numpy nd数组 ——> 张量： torch.tensor(nd数组)
    从标量中提取其内容： 标量张量.item()
"""
```

```python
import torch
import numpy as np
```

```python
# 1.演示：张量 ——> numpy

# 1.创建张量
t1=torch.tensor([1,2,3,4,5])
print(f't1: {t1}, type(t1): {type(t1)}')
# 2.张量 ——> numpy
n1=t1.numpy()
print(f'n1: {n1}, type(n1): {type(n1)}')

# 3.演示上述方式 共享内存
n1[0]=100
print(f'n1: {n1}')     # [100   2   3   4   5]
print(f't1: {t1}')     # [100,   2,   3,   4,   5]
```

```python
# 2.演示： numpy nd数组 ——> 张量

# 1.创建numpy数组
n1=np.array([1,2,3,4,5])
print(f'n1: {n1},type(n1): {type(n1)}')

# 2.numpy nd数组 ——> 张量
# t1=torch.from_numpy(n1).type(torch.float32)  # 转换 + 转类型
t1=torch.from_numpy(n1)
print(f't1: {t1}, type(t1): {type(t1)}')
t2=torch.tensor(n1)
print(f't2: {t2}, type(t2): {type(t2)}')

# 3.演示上述方式 是否共享内存
n1[0]=100
print(f'n1: {n1}')       # [100   2   3   4   5]
print(f't1: {t1}')       # [100   2   3   4   5]
print(f't2: {t2}')       # [1   2   3   4   5]
```

```python
# 3.演示：标量中提取其内容

# 1.创建张量
t1=torch.tensor(100.3)
print(f't1: {t1}, type(t1): {type(t1)}')

# 2.从张量中提取内容
a=t1.item()
print(f'value: {a}, type(a): {type(a)}')
```
